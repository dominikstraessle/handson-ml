# Vektoren
## Skalar
>Eine Grösse, die durch eine Zahl vollständig beschrieben ist, heisst Skalar.

In der Vektorrechnung ist ein Skalar einer reelen Zahl gleichzustellen.
## Vektor
>Eine Grösse, zu deren vollständiger Beschreibung neben einer Zahl noch die Angabe ihrer Richtung und Orientierung erforderlich ist, heisst Vektor.

**orientierte Strecke**: Eine Strecke mit Anfangs- und Endpunkt, wird durch einen Pfeil dargestellt.

**Schreibweise**: $\vec{a}$ -> gesprochen als "Vektor a"

```python
# Streckenabtragung
# addiere 18 * vektor u zu punkt p
p = np.array([1,3,-2])
u = np.array([7,4,4])
```

```python
u_length = np.linalg.norm(u)
amount = 18
```

```python
p_new = p + amount * (u / u_length)
print(p_new)
```

```python
# abstand zwischen zwei punkten
# 1. verbindungsvektor zwischen zwei punkten => endpunkt - anfangspunkt
# 2. Länge des verbindungsvektors
p_1 = np.array([7,4,2])
p_2 = np.array([3,7,2])
# 1.
gap = p_2 - p_1
# 2.
np.linalg.norm(gap)
```

```python
import math
```

```python
# winkel zwischen zwei vektoren
# 1. skalarprodukt berechnen
# 2. längen der vektoren berechnen
# 3. in formel einsetzen und nach winkel auflösen
u = np.array([2,2,1])
v = np.array([-1,-1,1])

skalar = v.dot(u)
u_length = np.linalg.norm(u)
v_length = np.linalg.norm(v)
radian = math.acos(skalar / (u_length * v_length))# gibt den wert im bogenmass zurück
math.degrees(radian)# lösung im gradmass
```

```python
def plot_vector2d(vector2d, origin=[0,0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
                     head_width=0.2, head_length=0.3, length_includes_head=True, **options)
```

```python
plot_vector2d(u, color='g')
plot_vector2d(v, color='b')
plt.axis([0,7,0,5])
plt.grid()
plt.show()
```

```python
def plot_vectors3d(ax, vectors3d, z0, **options):
    for v in vectors3d:
        x,y,z = v
        ax.plot([x,x], [y,y], [z0,z], color='grey', linestyle='dotted', marker='.')
    x_co, y_co, z_co = zip(*vectors3d)
    ax.scatter(x_co, y_co, z_co, **options)
```

```python
fig=plt.figure(figsize=(13,12), dpi= 60, facecolor='w', edgecolor='k')
subplot3d = fig.add_subplot(111, projection='3d')
subplot3d.set_zlim([0, 9])
plot_vectors3d(subplot3d, [a,b], 0, color=("r","b"))
plt.show()
```

```python
u = np.array([0,1])
v =np.array([1,0])
def vector_angle(u,v):
    cos_alpha = u.dot(v)/(np.linalg.norm(u) * np.linalg.norm(v))
    return np.degrees(np.arccos(cos_alpha))
print(vector_angle(u,v))
```
